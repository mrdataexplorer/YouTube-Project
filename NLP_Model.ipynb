```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
#Reading in the dataset
df = pd.read_csv('USvideos.csv')
```

```python
df.head()
```

```python
# Loading json data

import json

with open('US_category_id.json') as json_data:
    d = json.load(json_data)
    print(d)
```

```python
#Joining json data with the dataset

with open("US_category_id.json") as f:
    categories = json.load(f)["items"]
cat_dict = {}
for cat in categories:
    cat_dict[int(cat["id"])] = cat["snippet"]["title"]
df['category_name'] = df['category_id'].map(cat_dict)
```

```python
#Finding null values
df.isnull().sum()
```

```python
#Cleaning data
df["description"] = df["description"].fillna(value="")
```

```python
df.isnull().sum()
```

```python
#Splitting tags
df['tags_listed'] = df['tags'].map(lambda x: x.split('|'))
```

```python
df.head()
```

```python
#Creating variable with description
df1 = df['description']
df1
```

```python
#Initializing Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer

cvec = CountVectorizer(stop_words='english', max_features=1000)
```

```python
#Fitting model & transform
cvec.fit(df1)
new_df = cvec.transform(df1)
new_df
```

```python
# Turning words into numerical data
df2  = pd.DataFrame(new_df.todense(),
                   columns=cvec.get_feature_names())
df2
```

```python
# Finding average amount of words in the description
df2.mean().sort_values(ascending=False)[:20]
```

```python
# Finding the number of words frequently used.
df2.sum().sort_values(ascending=False)[:10]
```

```python
#Creating data frame for title
df_title = df['title']
df_title
```

```python
#Count vectorizing the title
cvec = CountVectorizer(stop_words='english', max_features=1000)
```

```python
#Fitting and transforming
cvec.fit(df_title)
new_df1 = cvec.transform(df_title)
new_df1
```

```python
# Turning words into numerical data
df4  = pd.DataFrame(new_df1.todense(),
                   columns=cvec.get_feature_names())
df4
```

```python
# The average amount of times a word appears in the description
df4.mean().sort_values(ascending=False)[:10]
```

```python
# The frequency of the word to appear in the description
df4.sum().sort_values(ascending=False)[:10]
```
